```python
# 4. View Results
import json
import pandas as pd
import matplotlib.pyplot as plt
import os

if os.path.exists('reports/fl_history.json'):
    with open('reports/fl_history.json', 'r') as f:
        history = json.load(f)

    rounds = [x[0] for x in history['accuracy']]
    accs = [x[1] for x in history['accuracy']]

    plt.plot(rounds, accs, marker='o')
    plt.title('Global Model Accuracy vs Rounds')
    plt.xlabel('Round')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.show()
else:
    print("Error: Training hasn't finished yet or reports/fl_history.json was not created. Please run Cell 3 first!")
```
